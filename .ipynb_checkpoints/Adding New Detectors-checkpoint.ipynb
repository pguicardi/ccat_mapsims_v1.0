# How to Add New Detectors with Custom Parameters

To add detectors to the mapsims blueprint, one must first go to $\rightarrow$ mapims $\rightarrow$ data $\rightarrow$ simonsobs_instrument_parameters_2020.06 $\rightarrow$ simonsobs_instrument_parameters_2020.06.tbl 

Detectors must be added in even numbers since simulations and information processing within the module is done in 2's. If you have an odd number, add a duplicate to the last spot and exclude it when analyzing any data. For example, the detectors HF5 and HF6 are identical and any simulations outputted by HF6 are excluded. Hence, if you wish to add an odd number of detectors you can just replace HF6 with one of these channels.  

You should see:



Add the detectors to this file in addition to the corresponding information. 

Additionally, in the same folder as this file, you should find bandpass data for each channel band. Copy and change the files for them to correspond to the newly added detectors. 



Now, go to $\rightarrow$ ccat_models $\rightarrow$ ccat_noise.py $\rightarrow$ class CcatLatv2b(SOLatType). Add the corresponding information to each attribute with a line of asterisks ("*****************") next to it.


!NOTE!: Make sure the channel tag and channel tube start with "LC" as separation between CCAT and SO channels is made based on this tag, which indicates it is a CCAT channel. Also, make sure the channel band starts with "HF." Specifically, make sure the bandpass file name is of the form: "bandpass_HF*.tbl". The counting of files of this form is important for the formatting of some behind-the-scenes features. 



After all these steps are finished, you should be able to run simulations of the new noise channels just as you would with any of the others.
